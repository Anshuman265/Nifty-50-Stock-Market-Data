```python
import opendatasets as od
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
file_paths = []
for root, directories, files in os.walk("archive"):
  for filename in files:
    filepath = os.path.join(root, filename)
    file_paths.append(filepath)
```

Code taken from :- https://www.geeksforgeeks.org/working-zip-files-python/

```python
stock_data = []
for i in range(len(file_paths)):
  file = (file_paths[i])
  if(file_paths[i]!="/archive/stock_metadata.csv"):
    stock_data.append(pd.read_csv(file))
  else:
    metadata = pd.read_csv(file)
```

```python
for i in range(0,len(stock_data)):
  c = stock_data[i].columns.get_loc("Symbol")
  #print(f'For i={i} and c={c} the data')
  #print(stock_data[i].iloc[0,c])
#  print(stock_data[i].iloc[-1,c])
  if(len(stock_data[i]) != 0):
    if(stock_data[i].iloc[0,c]!=stock_data[i].iloc[-1,c]):
      stock_data[i] = stock_data[i].assign(Symbol = stock_data[i].iloc[-1,c])
```
